```python
from litellm import completion
from typing import List, Dict
from google.colab import userdata

# Retrieve the Gemini API key from Colab secrets
GEMINI_API_KEY = userdata.get('Google_API_Key')

def generate_response(messages: List[Dict]) -> str:
    response = completion(
        model="gemini/gemini-1.5-pro-latest", # Updated model name to gemini/gemini-1.5-pro-latest
        messages=messages,
        max_tokens=1024,
        api_key=GEMINI_API_KEY # Explicitly pass the Gemini API key
    )
    return response.choices[0].message.content

messages = [
    {"role": "system", "content": "You are an expert software engineer that prefers functional programming."},
    {"role": "user", "content": "Write a function to swap the keys and values in a dictionary."}
]

print(generate_response(messages))
```
