```python
!pip install litellm
!pip install google-generativeai

import os
import google.generativeai as genai
from litellm import completion
from typing import List, Dict

# Replace "YOUR_API_KEY" with your actual Google AI Studio API key
GOOGLE_API_KEY = "YOUR_API_KEY"
genai.configure(api_key=GOOGLE_API_KEY)

def generate_response(messages: List[Dict]) -> str:
    """Call LLM to get response using Google Gemini"""
    # Convert messages to the format expected by Gemini
    gemini_messages = []
    for msg in messages:
        if msg["role"] == "user":
            gemini_messages.append({"role": "user", "parts": [msg["content"]]})
        elif msg["role"] == "system":
            # Gemini typically doesn't have a direct 'system' role in generate_content
            # For simplicity, we'll treat system messages as user messages for now
            # A more sophisticated approach might prepend system message to first user message
            gemini_messages.append({"role": "user", "parts": [msg["content"]]})
        elif msg["role"] == "model":
            gemini_messages.append({"role": "model", "parts": [msg["content"]]})

    model = genai.GenerativeModel('gemini-pro') # Using gemini-pro for text generation
    response = model.generate_content(gemini_messages)
    return response.text

messages = [
    {"role": "system", "content": "You are an expert software engineer that prefers functional programming."},
    {"role": "user", "content": "Write a function to swap the keys and values in a dictionary."}
]

# Uncomment the following lines and run the cell after setting your API key
# response = generate_response(messages)
# print(response)
```
